```python
import pandas as pd
df = pd.read_csv("./venv/data/ssafy_dataset_news_2023.csv", 
                sep='|',
                dtype={4: str, 5: str}
                )
df
```

```python
df.dtypes
```

```python
# filtering

# 특정 형식에 맞게 변환 (잘못된 형식은 NaT로 처리)
df['published'] = pd.to_datetime(df['published'], format='%Y-%m-%d %H:%M', errors='coerce')

word = '애플'
filtered_df = df[df['title'].str.contains(word, case=False, na=False)]
# 일주일 단위로 데이터를 그룹화하여 주별로 묶기 (일주일 시작일은 월요일)
weekly_filtered_df = filtered_df.set_index('published').resample('W').apply(lambda x: x.sort_values(by='published'))

# 결과 출력
weekly_filtered_df
```

```python
import pandas as pd

# 'published' 열을 datetime 형식으로 변환 (잘못된 형식은 NaT로 처리)
df['published'] = pd.to_datetime(df['published'], errors='coerce')

# 특정 단어로 필터링된 데이터프레임
word = '애플'
filtered_df = df[df['title'].str.contains(word, case=False, na=False)]

# 일주일 단위로 데이터를 그룹화하여 주별로 묶기 (일주일 시작일은 월요일)
weekly_filtered_df = filtered_df.set_index('published').resample('W').apply(lambda x: pd.DataFrame(x).sort_values('published'))

# 결과 출력
weekly_filtered_df
```
